```python
# ==================================
# Импорт необходимых библиотек
# ==================================
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
import tensorflow_datasets as tfds
from PIL import Image, ImageOps
from keras import regularizers
from sklearn.model_selection import train_test_split

# ==================================
# Загрузка и предобработка данных
# ==================================
# Загрузка набора данных Stanford Dogs
ds = tfds.load("stanford_dogs", split=['train', 'test'])

# Преобразование в DataFrame
df_train = tfds.as_dataframe(ds[0])
df_test = tfds.as_dataframe(ds[1])

# Печать размера первой картинки в обучающем наборе
print(df_train.iloc[0]['image'].shape)

# Изменение размера изображений до 64x96
df_train_resized = []
df_test_resized = []

for i in range(len(df_train)):
    img = Image.fromarray(df_train.iloc[i]['image'])
    img = img.resize((64, 96), Image.Resampling.LANCZOS)
    df_train_resized.append(np.array(img))

df_train['image'] = df_train_resized

for i in range(len(df_test)):
    img = Image.fromarray(df_test.iloc[i]['image'])
    img = img.resize((64, 96), Image.Resampling.LANCZOS)
    df_test_resized.append(np.array(img))

# Печать размера измененной первой картинки
print(df_train.iloc[0]['image'].shape)
plt.imshow(Image.fromarray(df_train.iloc[0]['image']))

# ==================================
# Разделение на обучающую, валидационную и тестовую выборки
# ==================================
# Разделение тестового набора на тестовый и валидационный (каждый по 25% от исходного тестового набора)
test_size = int(0.5 * len(df_test))
df_test, df_val = df_test[:test_size], df_test[test_size:]
```

```python
df_train.shape, df_test.shape, df_val.shape
```

```python
# Create empty DataFrames
df_train2 = pd.DataFrame()
df_test2 = pd.DataFrame()
df_val2 = pd.DataFrame()

# Select desired classes
selected_classes = [51, 53, 57, 58, 59]

# Filter data based on selected classes
df_train2 = df_train[df_train['label'].isin(selected_classes)]
df_test2 = df_test[df_test['label'].isin(selected_classes)]
df_val2 = df_val[df_val['label'].isin(selected_classes)]

# Display shapes of filtered data
df_train2.shape, df_test2.shape, df_val2.shape
```

```python
df_train2['label'].value_counts()
```

```python
# Function to plot random sample of images
def plot_random_sample(images):
    n = 10
    imgs = random.sample(list(images), n)

    num_row = 2
    num_col = 5

    fig, axes = plt.subplots(num_row, num_col, figsize=(3.5 * num_col, 3 * num_row))
    # For every image
    for i in range(num_row * num_col):
        # Read the image
        img = imgs[i]
        # Display the image
        ax = axes[i // num_col, i % num_col]
        ax.imshow(img)

    plt.tight_layout()
    plt.show()
```

```python
# Convert labels to numpy arrays
train_labels2 = df_train2['label'].to_numpy(dtype=np.float32)
test_labels2 = df_test2['label'].to_numpy(dtype=np.float32)
val_labels2 = df_val2['label'].to_numpy(dtype=np.float32)
train_labels2.shape, test_labels2.shape, val_labels2.shape
```

```python
# Initialize arrays for images
train_images2 = np.zeros(shape=(df_train2.shape[0], 96, 64, 3), dtype=np.float32)
test_images2 = np.zeros(shape=(df_test2.shape[0], 96, 64, 3), dtype=np.float32)
val_images2 = np.zeros(shape=(df_val2.shape[0], 96, 64, 3), dtype=np.float32)
train_images2.shape, test_images2.shape, val_images2.shape
```

```python
# Resize and assign images to arrays
for idx in range(train_labels2.shape[0]):
    img = Image.fromarray(df_train2.iloc[idx]['image']).resize((64, 96))
    train_images2[idx, :, :, :] = np.array(img)

for idx in range(test_labels2.shape[0]):
    img = Image.fromarray(df_test2.iloc[idx]['image']).resize((64, 96))
    test_images2[idx, :, :, :] = np.array(img)

for idx in range(val_labels2.shape[0]):
    img = Image.fromarray(df_val2.iloc[idx]['image']).resize((64, 96))
    val_images2[idx, :, :, :] = np.array(img)

# Display shapes of arrays
train_images2.shape, test_images2.shape, val_images2.shape
```

```python
train_images2 /= 255
test_images2  /= 255
val_images2 /= 255
```

```python
import random
plot_random_sample(test_images2)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization
```

```python
inputs = tf.keras.layers.Input(shape=(96, 64, 3))
```

```python
inputs.shape, inputs.dtype
```

```python
x = tf.keras.layers.Flatten()(inputs)
```

```python
# Adicionando as camadas ocultas
x = tf.keras.layers.Flatten()(inputs)
x = tf.keras.layers.Dense(120, kernel_initializer='lecun_normal', activation='selu', name='HiddenLayer1')(x)
x = tf.keras.layers.Dense(105, kernel_initializer='lecun_normal', activation='selu', name='HiddenLayer2')(x)
x = tf.keras.layers.Dense(90, kernel_initializer='lecun_normal', activation='selu', name='HiddenLayer3')(x)
x = tf.keras.layers.Dropout(0.2)(x)
x = tf.keras.layers.Dense(75, kernel_initializer='lecun_normal', activation='selu', name='HiddenLayer4')(x)
x = tf.keras.layers.Dense(60, kernel_initializer='lecun_normal', activation='selu', name='HiddenLayer5')(x)
x = tf.keras.layers.Dense(5, kernel_initializer='lecun_normal', activation='selu', name='HiddenLayer6')(x)

# Definindo a camada de saída
outputs = tf.keras.layers.Dense(5, activation='softmax')(x)

# Criando o modelo
mlp_model = tf.keras.Model(inputs=inputs, outputs=outputs)

# Exibindo o resumo do modelo
mlp_model.summary()
```

```python
tf.keras.utils.plot_model(mlp_model, show_shapes=True)
```

```python
mlp_model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer="sgd",
    metrics=["accuracy"],
)

callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history_1 = mlp_model.fit(
    train_images2,  # Corrected variable name
    train_labels2,
    epochs=100,
    batch_size=47,
    validation_data=(val_images2, val_labels2),
    callbacks=[callback]
)
```
